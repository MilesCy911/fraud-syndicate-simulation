```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()
```

```python
user_data = []

for user_id in range(2001, 2101):  # 100 users
    email_provider = random.choice(["yahoo.com", "gmail.com", "outlook.com", "aol.com"])
    country = random.choice(["Sweden", "Germany", "South Africa", "Norway"])
    email = fake.user_name() + "@" + email_provider
    reg_date = fake.date_between(start_date="-3y", end_date="today")
    deposit = round(random.uniform(100, 1000), 2)
    withdrawal = round(random.uniform(0, deposit), 2)
    logins = random.randint(1, 15)
    device_id = fake.uuid4()
    is_flagged = 0

    user_data.append([
        user_id, email, country, reg_date, deposit, withdrawal, logins, device_id, is_flagged
    ])
```

```python
df_faker = pd.DataFrame(user_data, columns=[
    "user_id", "email", "country", "registration_date",
    "deposit_amount", "withdrawal_amount", "login_count",
    "device_id", "is_flagged"
])

df_faker.head()
```

```python
df_faker.to_csv("user_transactions_100.csv", index=False)
```
